```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

## 1.Pie Chart

```python
x=delivery[delivery['batsman']=='V Kohli']
```

```python
y=x['batsman_runs'].value_counts()
y
```

```python
plt.pie(y,labels=y.index,autopct="%0.2f%%")
plt.show()
```

```python
def pie(name):
    x=delivery[delivery['batsman']==name]
    y=x['batsman_runs'].value_counts()
    
    plt.pie(y,labels=y.index,autopct="%0.2f%%")
    plt.show()  
pie('RG Sharma')
```

```python
plt.style.use('ggplot')
```

# 2. Line Chart

```python
big=match.merge(delivery,left_on='id',right_on='match_id')
```

```python
batsman_data=big[big['batsman']=='V Kohli']
batsman_data1=big[big['batsman']=='RG Sharma']
```

```python
print(batsman_data.groupby('season').sum()['batsman_runs'])
print(batsman_data1.groupby('season').sum()['batsman_runs'])
```

```python
performance=batsman_data.groupby('season').sum()['batsman_runs']
performance1=batsman_data1.groupby('season').sum()['batsman_runs']
performance
performance1
plt.plot(performance.index,performance.values,color="magenta",marker='s',markersize=10,label="Virat Kohli")
plt.plot(performance1.index,performance1.values,color="blue",marker='o',markersize=10,label="Rohit Sharma")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title("V Kohli & RG Sharma Season by Season Performance")
plt.legend()
```

```python
def comparator(batsman1,batsman2):
    big=match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data=big[big['batsman']==batsman1]
    batsman_data1=big[big['batsman']==batsman2]

    performance=batsman_data.groupby('season').sum()['batsman_runs']
    performance1=batsman_data1.groupby('season').sum()['batsman_runs']
    plt.plot(performance.index,performance.values,color="magenta",marker='s',markersize=10,label=batsman1)
    plt.plot(performance1.index,performance1.values,color="blue",marker='o',markersize=10,label=batsman2)
    plt.xlabel('Season')
    plt.ylabel('Runs Scored')
    plt.title("Batsman Season by Season Performance")
    plt.legend()
```

```python
comparator("AB de Villiers","CH Gayle")
```
